**Don't just copy and paste, actually type in the commands to load the data, and think about what each part of the command is telling the Python to do**


```python
# Load the CSV files and unpack
# NA data

nam_date, nam_data = np.loadtxt("North_America_living_index.csv", usecols=(0,1),unpack=True,skiprows=1,delimiter=',')
# freshwater data
fresh_date, fresh_data = np.loadtxt("Fresh_Water_living_index.csv", usecols=(0,1),unpack=True,skiprows=1,delimiter=',')
# asia pacific data
asia_date, asia_data = np.loadtxt("Asia_Pacific_Index.csv", usecols=(0,1),unpack=True,skiprows=1,delimiter=',')
```

```python
# verify that it imported as a NumPy array using the 'type' query

print(type(nam_data))
print(type(fresh_data))
print(type(asia_data))
```

With data imported into Python, we can begin to use NumPy's statistal ability to
determine some basic informations such as the mean, median, and standard deviation. To determine how to do this, check out the NumPy documentation here:
[statistics operations](https://docs.scipy.org/doc/numpy-1.13.0/reference/routines.statistics.html)

```python
# Type your code here

np.mean(nam_data)
np.median(nam_data)
np.std(nam_data)
```

Let's see if a graph of the data puts these values in better perspective.
To do that we will plot the date versus the data using matplotlib. Do this for the three files so they are all on one graph.
This is how data science begins to come alive; using real data and making sense of the values. 
While it is a simple, first step. Understanding how data can be presented and represented (or misrepresented) are important concepts.

```python
# Type your code here

plt.plot(nam_date, nam_data, label='North America')
plt.plot(fresh_date, fresh_data, label='Fresh Water')
plt.plot(asia_date, asia_data, label='Asia Pacific')

plt.title('Global Wildlife Changes Over Time')
plt.xlabel('Year')
plt.ylabel('Index Value')
plt.legend()
plt.show()
```


_Make sure your graph above has a title, labels and legend._
